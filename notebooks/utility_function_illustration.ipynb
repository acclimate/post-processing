```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# two good illustration

#share of good 1 vs good 2

def two_good_CES_utility (share_good_1,sigma=2,share_factor=0.5):
    return (share_factor**(1/sigma) * share_good_1**((sigma-1)/sigma) + (1-share_factor)**(1/sigma) * (1-share_good_1)**((sigma-1)/sigma))**(sigma/(sigma-1))


# 100 linearly spaced numbers
x = np.linspace(0,1,100)

# the function, which is y = x^3 here
y = two_good_CES_utility(x,sigma=0.5)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.ylim(0,1)
plt.rcParams['text.usetex'] = True
sigmas = [0.125,0.25,0.5,0.75,1.25,2,3,6.25,8]
for i_sigma in sigmas:
    y = two_good_CES_utility(x,sigma=i_sigma)
    plt.plot(x,y,label=r'$\sigma=$'+r'${}$'.format(str(i_sigma)))
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.savefig("CES_illustration.pdf", bbox_inches='tight')
```
